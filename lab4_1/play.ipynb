```python
import numpy as np
import matplotlib.pyplot as plt

#    x
x = np.linspace(-10, 10, 400)

#  
y1 = 2 * x + 1   # y < 2x + 1
y2 = -0.5 * x + 3  # y > -0.5x + 3
y3 = x**2 - 4  # y < x^2 - 4

#  
plt.figure(figsize=(8, 6))

#  ,  
plt.fill_between(x, -10, y1, where=(y1 > -10), color='lightblue', alpha=0.5, label='y < 2x + 1')
plt.fill_between(x, y2, 10, where=(y2 < 10), color='lightgreen', alpha=0.5, label='y > -0.5x + 3')
plt.fill_between(x, -10, y3, where=(y3 > -10), color='salmon', alpha=0.5, label='y < x - 4')

#   
plt.xlim(-10, 10)
plt.ylim(-10, 10)

#  
plt.title(' ')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='dashed')
plt.axvline(0, color='black',linewidth=0.5, ls='dashed')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
